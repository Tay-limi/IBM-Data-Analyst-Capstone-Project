# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT ConvertedComp
FROM master
"""

df_ConvertedComp = pd.read_sql_query(QUERY, conn)
df_ConvertedComp.head()
```

```python
# replace null values with the mean of the ConvertedComp
import numpy as np
mean = df_ConvertedComp['ConvertedComp'].mean()
mean = round(mean)
df_ConvertedComp['ConvertedComp'].replace(np.nan, mean, inplace = True)
```

```python
# plotting the Histogram
import matplotlib.pyplot as plt
import numpy as np

count, bin_edges = np.histogram(df_ConvertedComp, bins = 10)
plt.figure(figsize =(10,6))
plt.hist(x = df_ConvertedComp, bins = 10, edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel("Salary(USD)")
plt.ylabel("Count")
plt.xticks(bin_edges)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# get the age column
QUERY = """
SELECT Age
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df.head()
```

```python
# replace nan values with mean age
mean_age = df['Age'].mean()
mean_age = round(mean_age)
df['Age'].replace(np.nan, mean_age, inplace=True)
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# get the columns
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

#clean the data
# replace nan values with mean age and mean WorkWeekHrs
mean_age = df['Age'].mean()
mean_age = round(mean_age)
df['Age'].replace(np.nan, mean_age, inplace=True)

mean_hrs = df['WorkWeekHrs'].mean()
mean_hrs = round(mean_hrs)
df['WorkWeekHrs'].replace(np.nan, mean_hrs, inplace=True)

# scatter plot
plt.figure(figsize=(10,6))
plt.scatter(x=df['WorkWeekHrs'], y=df['Age'])
plt.title("Relationhsip between Age and Work Week Hours")
plt.xlabel("Work Week Hours")
plt.ylabel("Age")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# get the columns
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
df.head() 

# replace nan values with mean age, mean WorkWeekHrs and mean CodeRehrs
mean_age = df['Age'].mean()
mean_age = round(mean_age)
df['Age'].replace(np.nan, mean_age, inplace=True)

mean_hrs = df['WorkWeekHrs'].mean()
mean_hrs = round(mean_hrs)
df['WorkWeekHrs'].replace(np.nan, mean_hrs, inplace=True)

mean_chrs = df['CodeRevHrs'].mean()
mean_chrs = round(mean_hrs)
df['CodeRevHrs'].replace(np.nan, mean_chrs, inplace=True)

# bubble plot
import seaborn as sns
size = df['Age']
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="CodeRevHrs", y="WorkWeekHrs", size=size, hue='Age')
plt.title("Relationship between Work week hrs and Code Review Hours")
plt.ylabel(" Work week hrs")
plt.xlabel("Code Review Hours")
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# get the top 5 databases
QUERY= """
SELECT *, Count(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count Desc
LIMIT 5
"""
df = pd.read_sql_query(QUERY, conn)
df
```

```python
# plot the pie chart
labels = df['DatabaseDesireNextYear'].values
plt.figure(figsize=(10,6))
plt.pie(x = df['Count'], autopct='%1.1f%%', startangle=90, labels=labels)
plt.title("Top 5 databases Desired for the following year")
plt.legend()
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age between 30 and 35
"""

df = pd.read_sql_query(QUERY, conn)
median_Work_hrs = df['WorkWeekHrs'].median()
median_Code_hrs = df['CodeRevHrs'].median()

#Plot the stacked bar
plt.figure(figsize=(10,6))
plt.bar('Median',median_Work_hrs, label='Work Week Hrs')
plt.bar('Median',median_Code_hrs, label = 'Code Rev Hrs')
plt.ylabel('Median Hours')
plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age
from master
WHERE Age between 45 and 60
"""
df = pd.read_sql_query(QUERY, conn)

#replace the null values in the ConvertedComp column with its mean value
mean = df['ConvertedComp'].mean()
mean = round(mean)
df['ConvertedComp'].replace(np.nan, mean, inplace = True)

# Calculate median ConvertedComp for each age
median_comp_by_age = df.groupby('Age')['ConvertedComp'].median()

# Plot the median ConvertedComp for each age
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch from 
master
"""
df = pd.read_sql_query(QUERY, conn)

# Count the occurrences of each unique value in the MainBranch column
main_branch_counts = df['MainBranch'].value_counts()

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Count of MainBranch')
plt.grid(axis='x')
plt.show()
```
